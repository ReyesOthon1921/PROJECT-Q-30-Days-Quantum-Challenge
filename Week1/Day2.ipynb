# What Is a Qubit?

A qubit (quantum bit) is the fundamental unit of information in a quantum computer.

Unlike a classical bit, which stores either **0** or **1**, a qubit is described by the rules of quantum mechanics.

Before measurement, a qubit is represented by a quantum state rather than a single classical value.

This allows quantum computers to process information differently from classical computers.

---

# Computational Basis States

A single qubit uses two computational basis states:

* **|0⟩**
* **|1⟩**

These basis states provide the reference points used to describe quantum information.

After measurement, the qubit collapses into one of these basis states.

---

# Probability Before Measurement

One important idea I continued developing today is that quantum computing does not describe a qubit using a definite value before measurement.

Instead, the quantum system is described mathematically, and measurement determines which classical outcome is observed.

This idea prepares the foundation for learning about quantum state vectors and probability amplitudes in the next lesson.

---

# Measurement

Measurement is the process of converting quantum information into classical information.

Although the internal quantum state may contain multiple possibilities before measurement, the observed result is always one classical value.

For a single qubit, that outcome is either:

* **|0⟩**
* **|1⟩**

Understanding measurement is one of the key ideas required before studying quantum algorithms.

---

# Why This Matters

The concepts learned today explain why quantum computing requires mathematics that differs from classical computing.

Rather than assigning fixed values to information, quantum systems describe possible outcomes using mathematical models that evolve until measurement occurs.

This provides the foundation for understanding state vectors, amplitudes, probability, and quantum gates.
